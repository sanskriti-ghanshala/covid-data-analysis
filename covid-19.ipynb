```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Dell\Desktop\covid-19ds.csv")
```

```python
df.head()
```

```python
top_cases = df.sort_values("Confirmed", ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x="Confirmed", y="Country/Region", data=top_cases)

plt.title("Top 10 Countries with Highest COVID Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("Country")

plt.show()
```

```python
top_deaths = df.sort_values("Deaths", ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x="Deaths", y="Country/Region", data=top_deaths)

plt.title("Top 10 Countries with Highest Deaths")
plt.xlabel("Deaths")
plt.ylabel("Country")

plt.show()
```

```python
top_recovered = df.sort_values("Recovered", ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x="Recovered", y="Country/Region", data=top_recovered)

plt.title("Top 10 Countries with Highest Recoveries")
plt.xlabel("Recovered Cases")
plt.ylabel("Country")

plt.show()
```

```python
top_active = df.sort_values("Active", ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x="Active", y="Country/Region", data=top_active)

plt.title("Top 10 Countries with Active Cases")
plt.xlabel("Active Cases")
plt.ylabel("Country")

plt.show()
```

```python
region_cases = df.groupby("WHO Region")["Confirmed"].sum()

region_cases.plot(kind="bar", figsize=(8,5))

plt.title("COVID Cases by WHO Region")
plt.xlabel("Region")
plt.ylabel("Total Cases")

plt.show()
```

```python
numeric_data = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))

sns.heatmap(numeric_data.corr(), annot=True)

plt.title("COVID Dataset Correlation Heatmap")

plt.show()
```
